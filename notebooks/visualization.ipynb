```python
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt

from config.settings import (
    DB_HOST,
    DB_PORT,
    DB_NAME,
    DB_USER,
    DB_PASSWORD
)
```

```python
connection = psycopg2.connect(
    host=DB_HOST,
    port=DB_PORT,
    database=DB_NAME,
    user=DB_USER,
    password=DB_PASSWORD
)

print(" Connected to PostgreSQL")
```

```python
query = """
SELECT
    trade_date,
    close,
    volume,
    sma_20,
    ema_12,
    rsi_14,
    macd,
    macd_signal,
    bb_upper,
    bb_lower,
    obv
FROM processed_stock_data
WHERE company_id = 1
ORDER BY trade_date;
"""

df = pd.read_sql(query, connection)

df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df["trade_date"], df["close"], label="Close Price")
plt.plot(df["trade_date"], df["sma_20"], label="SMA 20")
plt.plot(df["trade_date"], df["ema_12"], label="EMA 12")

plt.title("Apple Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.plot(df["trade_date"], df["rsi_14"], label="RSI", color="purple")

plt.axhline(70, linestyle="--", color="red")
plt.axhline(30, linestyle="--", color="green")

plt.title("Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.plot(df["trade_date"], df["macd"], label="MACD")
plt.plot(df["trade_date"], df["macd_signal"], label="Signal")

plt.title("MACD Indicator")
plt.xlabel("Date")
plt.ylabel("MACD")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df["trade_date"], df["close"], label="Close")
plt.plot(df["trade_date"], df["bb_upper"], label="Upper Band")
plt.plot(df["trade_date"], df["bb_lower"], label="Lower Band")

plt.title("Bollinger Bands")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df["trade_date"], df["obv"])

plt.title("On Balance Volume (OBV)")
plt.xlabel("Date")
plt.ylabel("OBV")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.bar(df["trade_date"], df["volume"])

plt.title("Daily Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.grid(True)

plt.show()
```

```python
correlation = df.drop(columns=["trade_date"]).corr()

correlation
```

```python
plt.figure(figsize=(10,8))

plt.imshow(correlation, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)

plt.title("Feature Correlation Matrix")

plt.show()
```

```python
connection.close()

print(" Database Connection Closed")
```
